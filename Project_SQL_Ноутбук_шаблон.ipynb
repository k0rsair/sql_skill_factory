```python
# текст запроса

query_5_1 = f'''with employer_cnt as (
                select emp.id,
                emp.name,
                count(vac.*) as vac_cnt,
                dense_rank() over (order by count(vac.*) desc) as vac_rank
                from employers emp
                join vacancies vac
                on vac.employer_id = emp.id
                group by emp.id, emp.name
            )
            select id, name, vac_cnt, vac_rank
            from employer_cnt
            where vac_rank in (1, 5)
            order by vac_rank, name
'''
```

```python
# код для получения списка городов-милионников
import requests
from bs4 import BeautifulSoup

# URL страницы Википедии
url = "https://ru.wikipedia.org/wiki/Города-миллионеры_России"
headers = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# Делаем запрос к странице
response = requests.get(url, headers=headers, timeout=10)
response.raise_for_status()
# Парсим HTML
soup = BeautifulSoup(response.text, "html.parser")

# Находим таблицу со списком городов-миллионников
table = soup.find("table")

cities = []

# Итерируем строки таблицы
for row in table.find_all("tr")[1:]:
    cols = row.find_all("td")
    if cols:
        city_name = cols[1].get_text(strip=True)
        cities.append(city_name)

cities
```

```python
# текст запроса
cities_joined = ", ".join(f"'{city}'" for city in cities)

query_5_7 = f'''WITH million_cities AS (
    SELECT id, name
    FROM areas
    WHERE name IN (
        {cities_joined}
    )
),
yandex_vacancies AS (
    SELECT mc.name AS city,
           COUNT(v.id) AS cnt
    FROM vacancies v
    JOIN employers e ON e.id = v.employer_id
    JOIN million_cities mc ON mc.id = v.area_id
    WHERE e.name = 'Яндекс'
    GROUP BY mc.name
)

SELECT city, cnt
FROM (
    SELECT city, cnt
    FROM yandex_vacancies

    UNION ALL

    SELECT 'Total' AS city, SUM(cnt) AS cnt
    FROM yandex_vacancies
) t
ORDER BY cnt ASC'''
```

6. Напишите запрос, позволяющий вычислить, какую зарплату для DS в среднем указывают для каждого типа требуемого опыта (уникальное значение из поля `experience`).

    При решении задачи примите во внимание следующее:
    1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
    2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
    3. Если в расчётах участвует `null`, в результате он тоже даст `null` (посмотрите, что возвращает запрос `select 1 + null`). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит `null` на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

    Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до **целого числа**.

```python
# текст запроса
query_6_5 = f'''SELECT experience,
             ROUND(
                AVG(case
                when
                    salary_from is not null and salary_to is not null then
                        (salary_from + salary_to) / 2
                else COALESCE(salary_from, 0) + COALESCE(salary_to, 0)
                end)) AS avg_salary
            FROM vacancies
            {ds_where}
            AND (salary_from IS NOT NULL OR salary_to IS NOT NULL)
            GROUP BY experience'''
```

```python
# результат запроса
get(query_6_5)
```
